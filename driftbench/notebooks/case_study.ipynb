```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 设置 seaborn 风格
sns.set(style="whitegrid")
plt.rcParams.update({'font.size': 12})

# Baselines
baselines = ['PostgreSQL', 'Naru', 'MSCN']
colors = sns.color_palette("Set2", len(baselines))

# Q-error 数据
q00 = [1.3, 1.2, 1.1]
q01 = [3.2, 2.1, 1.1]
q02 = [4.5, 2.5, 2.2]
q11 = [6.1, 3.0, 3.0]

data = [q00, q01, q02]
titles = [
    "Original Data + Original Query",
    "Original Data + Query Drift",
    "Data Drift + Original Query",
    # "Data Drift + Query Drift"
]

fig, axes = plt.subplots(1, 3, figsize=(8, 3))
width = 0.6
x = range(len(baselines))
ymax = max(max(d) for d in data) * 1.3

for idx, ax in enumerate(axes.flat):
    bars = ax.bar(x, data[idx], width, color=colors)

    # 添加数值标注
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, f"{yval:.1f}",
                ha='center', va='bottom', fontsize=11)

    ax.set_xticks(x)
    ax.set_xticklabels(baselines, fontsize=11)
    ax.set_title(titles[idx], fontsize=12, weight='bold')
    ax.set_ylim(0, ymax)
    ax.set_ylabel("Q-error", fontsize=11)

plt.tight_layout()

output_dir = "case_study"
overleaf_dir = "/Users/guanlil1/Dropbox/应用/Overleaf/DataWorkloadGen/figures/"
file_name="case_study"
# plt.suptitle("Q-error under Data and Query Drift", fontsize=16, weight='bold', y=1.05)
plt.savefig(os.path.join(output_dir, f"{file_name}.png"), format='png', bbox_inches='tight')
plt.savefig(os.path.join(overleaf_dir, f"{file_name}.pdf"), format='pdf', bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import os
import pandas as pd

is_fig_save = True
overleaf_dir = "/Users/guanlil1/Dropbox/应用/Overleaf/DataWorkloadGen/figures/"
output_dir = "case_study"
os.makedirs(output_dir, exist_ok=True)

def read_q_error(csv_file):
    df = pd.read_csv(csv_file)
    return df["error"].tolist()


def get_data(data_files):
    res_dir = "../../existing_benchmarks/AreCELearnedYet/output/result/census13/"
    
    data = []
    for i in range(len(postgres)):
        data.append(read_q_error(os.path.join(res_dir, data_files[i])))
    return data
```

```python
import matplotlib.pyplot as plt

def plot_qerror_boxplot(data, labels, tag, file_name):
    plt.figure(figsize=(13, 4))


    # '#8da0cb', '#fc8d62', '#66c2a5'
    plt.boxplot(
        data, labels=labels, patch_artist=True,
        boxprops=dict(facecolor='lightblue', color='#8da0cb'),
        medianprops=dict(color='#fc8d62'),
        whiskerprops=dict(color='#66c2a5'),
        capprops=dict(color='#66c2a5'),
        flierprops=dict(
            marker='o',
            markersize=3,
            alpha=0.6,
            markerfacecolor='#fc8d62',
            markeredgecolor='#fc8d62'
        )
    )


    for i in range(6, len(labels), 6):
        plt.axvline(x=i + 0.5, color='gray', linestyle='--', linewidth=2, alpha=0.9)

    # 添加分组标题（D_1, D_2, D_3）
    group_titles = ["census", "census-drift1", "census-drift2"]
    group_centers = [3.5, 9.5, 15.5] 

    ylim = plt.ylim()
    for x, title in zip(group_centers, group_titles):
        plt.text(x, ylim[1] * 0.9, title, ha='center', va='bottom', fontsize=22)

    plt.xticks(rotation=0, fontsize=18)
    plt.yticks(fontsize=24)
    plt.ylabel("Q-error", fontsize=24)
    plt.title(f"Q-error Distribution Across {tag}", fontsize=24)
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.tight_layout()
    
    if is_fig_save and file_name:
        plt.savefig(os.path.join(output_dir, f"{file_name}.png"), format='png', dpi=300, bbox_inches='tight')
        plt.savefig(os.path.join(overleaf_dir, f"{file_name}.pdf"), format='pdf', bbox_inches='tight')
    plt.show()

plot_qerror_boxplot(get_data(postgres), labels, "PostgreSQL", "case_study_pg")
plot_qerror_boxplot(get_data(naru), labels, "Naru", "case_study_naru")
plot_qerror_boxplot(get_data(mscn), labels, "MSCN" , "case_study_mscn")
```
